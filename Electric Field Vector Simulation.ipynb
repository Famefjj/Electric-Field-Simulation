```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1st element for x and y
xt = np.linspace(-2,2);
yt = np.linspace(-2,2);

eo = 8.854187817*10**(-12);
x,y = np.meshgrid(xt,yt);

```

```python
Ex = x/(4*np.pi*eo*(x**2+y**2)**(3/2));
Ey = y/(4*np.pi*eo*(x**2+y**2)**(3/2));

# plt.quiver(x,y,Ex,Ey);
```

```python
# 2nd element
# xf = np.linspace(-5,5, 10);
# yf = np.linspace(-5,5, 10);

xs,ys = np.meshgrid(xt,yt);
xc = 1.0; #center
yc = 0;
Exs = -(x-xc)/(4*np.pi*eo*((x-xc)**2+(y-yc)**2)**(3/2));
Eys = -(y-yc)/(4*np.pi*eo*((x-xc)**2+(y-yc)**2)**(3/2));

# plot both elements
Exr = Ex + Exs;
Eyr = Ey + Eys;

fig = plt.figure(figsize= (8,8));
ax = fig.add_subplot(1,1,1, aspect=1);

color = 2*np.log(np.hypot(Ex, Ey));
ax.streamplot(xs, ys, Exr, Eyr, color=color,cmap=plt.cm.plasma, linewidth=1,density=2);
ax.add_patch(plt.Circle((0,0), 0.05, color='r'))
ax.add_patch(plt.Circle((xc,yc), 0.05, color='blue'))

plt.show()
```
